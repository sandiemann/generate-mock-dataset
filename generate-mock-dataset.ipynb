```python
mock = Faker() 
def create_custom_data(x): 
  
    # dictionary 
    custom_data ={} 
    for i in range(0, x): 
        custom_data[i]={} 
        custom_data[i]['name']= mock.name() 
        custom_data[i]['address']= mock.address() 
        custom_data[i]['city']= mock.city() 
        custom_data[i]['job']= mock.job()
        custom_data[i]['Hire Date'] = mock.date_between(start_date='-30y', end_date='today'),
    
    return custom_data
```

```python
user_data = create_custom_data(20)
pd.DataFrame.from_dict(user_data)
```
